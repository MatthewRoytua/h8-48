## Area Pots (Series/Dataframe) <a id="12"></a>

**What is a line plot and why use it?**

An Area chart or area plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

**Questions:**

With area plot, show 5 most criminal boroughs!

```python
# finding most criminal boroughs

top = criminal.groupby('borough')['value'].sum().sort_values(ascending = False) 
top_5 = top.head()
top_5
```

```python
# filtering data contained with the most criminal borough and setting new index with year

filter_list = ['Croydon', 'Lambeth', 'Ealing','Southwark','Barnet']
df_top5 = criminal[criminal.borough.isin(filter_list)]
df_top5.set_index('year', inplace = True)
df_top5
```

```python
# the amount of crime each borough has from 2008 - 2016

totals = df_top5.groupby('year')['borough'].value_counts()
df = totals.to_frame()
df
```

```python
# unstacking index and transposing data for plotting

df2 = df.unstack(level='year')
df2.columns = df2.columns.droplevel()
df3 = df2.transpose()
df3
```

```python
# drawing

df3.plot(kind='area', 
             stacked=False,
             alpha=0.5,
             figsize=(20, 10),
             )

plt.title('Top 5 Most Criminal Boroughs')
plt.ylabel('Number of Crimes')
plt.xlabel('Years')

plt.show()
```

**Insight:**

From the data, we can conclude that top 5 most criminal Boroughs relatively has consistent amount of crimes overyears.

## Histogram

A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

**Question:**

Using histogram, what is the frequency distribution of the number of cases from the various minor categories of crime happened in 2008?

```python
# sorting crime data in 2008

cr = criminal.loc[criminal.year == 2008]
cr
```

```python
# grouping based on minor category

cr_min = cr.groupby('minor_category')['value'].count()
cr_min
```

```python
# convert series into frame

df_cr = cr_min.to_frame()
df_cr
```

```python
# making bin edges

count, bin_edges = np.histogram(df_cr)

print(count)
print(bin_edges)
```

```python
# drawing
df_cr.plot(kind = 'hist', figsize = (8,5),edgecolor = 'black')


plt.title('Histogram of Minor Category of Crime Happened in 2008')
plt.ylabel('Number of Minor Category of Crime')
plt.xlabel('Number of Cases')

plt.show()
```

**Insight:**

* In 2008, there are 7 minor categories contributed between 0 to 1790.9 crimes 
* 4 minor categories contributed between 1790.9 to 3455.4 crimes, and so on...

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

Using Bar Charts, what year is the crime at its peak?

```python
# grouping based on year

bar = criminal.groupby('year')['value'].count()
bar.sort_values()
```

```python
# drawing

bar.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of Cases') # add y-label to the plot
plt.title('Criminal Cases from 2008 to 2016') # add title to the plot

plt.show()
```

## Pie Charts <a id="6"></a>

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media. We can create pie charts in Matplotlib by passing in the `kind=pie` keyword.

**Question:**

Using pie chart, explore the proportion (percentage) of cases grouped by major categories in 2008 up until 2016.

```python
# convert year to str to skip sum process

criminal['year'].astype(str)

# grouping based on majer category

pie = criminal.groupby('major_category')['value'].count()
pie
```

```python
# ratio for each continent with which to offset each wedge.

explode_list = [0, .1, 0, .7, .2, .1, .4, 0, .1]

# drawing

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#fcc603','#c2c2f0','#ffb3e6', 'yellowgreen','pink']
pie.plot(kind='pie',
         figsize=(15, 6),
         autopct='%1.1f%%', # add in percentages
         startangle=90,     # start angle 90° (Africa)
         shadow=True, 
         explode=explode_list,
         colors = colors,
         labels=None,         # turn off labels on pie chart
         pctdistance=1.12,     
        )

plt.title('All Major Criminal Categories', y=1.12)

# add legend
plt.legend(bbox_to_anchor=(0,1), labels=pie.index, loc='upper right') 

# donut is way better than pie!!!!!!!!
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle  
plt.tight_layout()
plt.show()
```

**Insight:**

Theft and Handling has 30.9% contribution, which makes it the most major category contributed to the total amount of crimes.

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

**Question:**

Create a box plot to visualize the distribution crimes happened in Croydon, Lambeth, and Southwark from 2008 - 2016!

```python
# filtering related data

filter_list = ['Croydon', 'Lambeth','Southwark']
box = criminal[criminal.borough.isin(filter_list)]
box
```

```python
# grouping data based on year, converting series to data frame, unstacking index and transposing  for plotting

boxs = box.groupby('year')['borough'].value_counts().to_frame()
df_box = boxs.unstack(level='year')
df_box.columns = df_box.columns.droplevel()
df_boxs = df_box.transpose()
df_boxs
```

```python
# drawing

df_boxs.plot(kind='box', figsize=(8, 6))

plt.title('Box plot of Croydon, Lambeth, and Southwark Crimes from 2008 - 2016')
plt.ylabel('Number of Crimes')

plt.show()
```

**Insight:**

* Southwark has the lowest crimes
* Lambeth has the longest whiskers relative to Croydon and Soutwark


## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**

Create a scatter plot of the total crimes from 2008 to 2016!

```python
# grouping based on year and resetting index to include year on variables

agg = criminal.groupby('year')['value'].count().to_frame()
agg.reset_index(inplace = True)
agg
```

```python
# drawing

agg.plot(kind='scatter', x = 'year', y= 'value', figsize=(10, 6), color='darkblue')

plt.title('Total of Crimes from 2008 to 2016')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')

plt.show()
```

## Word Clouds <a id="8"></a>


`Word` clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.

```python
# install wordcloud
# !conda install -c conda-forge wordcloud --yes

!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
stopwords = set(STOPWORDS)
```

```python
# grouping data based on borough, and converting it in to data frame

word = criminal.groupby('borough')['value'].count().to_frame()
word
```

```python
# making variable of total crimes

total = word['value'].sum()
total
```

```python
max_words = 50
word_string = ''
for borough in word.index.values:
    # check if country's name is a single-word name
    if len(borough.split(' ')) == 1:
        repeat_num_times = int(word.loc[borough, 'value']/float(total)*max_words)
        word_string = word_string + ((borough + ' ') * repeat_num_times)
                                     
# display the generated text
word_string
```

```python
# creating the word cloud

wordcloud = WordCloud(background_color='white').generate(word_string)

print('Word cloud created!')
```

```python
# display the cloud

fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(15)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
